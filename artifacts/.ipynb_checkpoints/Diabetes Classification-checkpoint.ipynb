## Outliers Detection

```python
df1 = df.drop('Outcome',axis = 1)
for i in df1.columns :
    plt.figure(figsize = (2,2))
    sns.boxplot(df[i])
    plt.show()
```

## Treating Outliers

```python
for i in df1.columns :
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3-q1
    lt = q1 - 1.5*iqr
    ut = q3 + 1.5*iqr
    
    df.loc[df[i]<lt,i] = lt
    df.loc[df[i]>ut,i] = ut
    
    plt.figure(figsize = (2,2))
    sns.boxplot(df[i])
    plt.show()
```

## Over Sampling (SMOTE)

```python
x = df.drop('Outcome',axis = 1)
y = df['Outcome']
```

```python
smote = SMOTE(sampling_strategy = 0.95)
x_re,y_re = smote.fit_resample(x,y)
```

## Logistic Regression

```python
x_train,x_test,y_train,y_test = train_test_split(x_re,y_re,random_state = 5,test_size = 0.3,stratify = y_re)
x_train.shape,x_test.shape,y_train.shape,y_test.shape 
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_proba = lr.predict_proba(x_test)
y_pred_proba[0,1]
```

## Log Odds

```python
log_odds = []
for i in range(len(y_pred_proba)) :
    odds = np.log(y_pred_proba[i,0]/y_pred_proba[i,1])
    log_odds.append(odds)

log = pd.DataFrame(log_odds,columns = ['log_odd'])
log
```

```python
def model_eval(x,y,model):
    y_pred = model.predict(x)
    cm = confusion_matrix(y,y_pred)
    print(f"Confusion Matrix = \n{cm}")
    
    clf = classification_report(y,y_pred)
    print(f"Classification Report = \n{clf}")
    
    recall = recall_score(y,y_pred)
    print(f"Recall = {recall}")
    
    precision = precision_score(y,y_pred)
    print(f"Precision = {precision}")
    
    acc = accuracy_score(y,y_pred)
    print(f"Accuracy Score = {acc}")
    
    return plot_confusion_matrix(cm)
```

```python
model_eval(x_test,y_test,lr)
```
